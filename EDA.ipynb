```python
# Create a bar plot for the 'cash_type' column
df.cash_type.value_counts().plot(kind='bar',color='orange')
plt.title("Amount")
```

```python
# Create a pie plot for the 'cash_type' column
df.cash_type.value_counts().plot(kind='pie')
plt.title("Money")
```

```python
# Create a pie plot for the 'coffee_name' column
df.coffee_name.value_counts().plot(kind='pie')
plt.title("Varieties of Coffee")
```

```python
# Set the figure size
plt.figure(figsize=(7, 4))

# Create a bar plot for coffee order frequency
df['coffee_name'].value_counts().plot(kind='bar', color='orange')

# Add a title to the plot
plt.title('Coffee Order Frequency')

# Show the plot
plt.show()
```

```python
# Set the figure size
plt.figure(figsize=(10, 6))

# Create a box plot
sns.boxplot(x='coffee_name', y='money', data=df, color='orange')

# Rotate the x-axis labels
plt.xticks(rotation=45, ha='right', fontsize=12)

# Add a title
plt.title('Money Spent by Coffee Type')

# Show the plot
plt.show()
```

```python
# Create a relational plot
sns.relplot(x='coffee_name', y='money', hue='cash_type', data=df)

# Rotate the x-axis labels
plt.xticks(rotation=45, ha='right', fontsize=12)

# Add a title
plt.title("Relationship between Coffee Type and Money Spent")

# Show the plot
plt.show()
```

```python
# Create a distribution plot for 'coffee_name'
sns.displot(df['coffee_name'],color='skyblue', kde=True, aspect=3)

# Add a title
plt.title('Distribution of dates in plot')

# Rotate the x-axis labels
plt.xticks(rotation=45, ha='right', fontsize=1)
```

```python
# Create a distribution plot for 'money'
sns.displot(df['money'],color='orange', kde=True)

# Add a title
plt.title('Collection of money in plot')
```

```python
# create a categorical plot for 'cash_type' and 'coffee_name'
sns.catplot(x='cash_type',y='coffee_name',kind='strip', color='orange', data = df)

# Add a title
plt.title('Strip Plot of Coffee Types by Date')
```
